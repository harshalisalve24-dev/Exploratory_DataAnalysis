 Why can't we directly use Machine Learning?

Imagine this dataset:
a1.PNG
Immediately, questions arise:

Why is one price missing?
Why is one price 9999?
Is it a luxury car or a data entry error?
What's the average price?
Which brand appears most?

Those are EDA questions.
 The EDA Mindset

Don't think:

"I need to use .mean()."

Instead think:

"What business question am I trying to answer?"

```python
import numpy as np

prices = np.array([450000, 600000, 750000, 500000, 900000])
print(type(prices))
```

```python
## Print the average priceimport numpy as np

prices = np.array([450000, 600000, 750000, 500000, 900000])
print(prices.mean())
```

```python
## Print the highest price
prices = np.array([450000, 600000, 750000, 500000, 900000])
print(prices.max())
```

```python

# Print the lowest price
prices = np.array([450000, 600000, 750000, 500000, 900000])
print(prices.min())
```

```python
## Print the number of cars
prices = np.array([450000, 600000, 750000, 500000, 900000])
print(len(prices))
```

```python
prices = np.array([450000, 600000, 750000, 500000, 900000])
prices = prices * 1.10
print(prices)
```
